# 🤖 AI Multi-Task Automation Agent (NLP-Based)

This project simulates an AI agent that can:
- Generate emails
- Summarize text
- Suggest code
- Classify user intent using NLP-style logic

```python
# If needed later for upgrades
# !pip install nltk
```

```python
def detect_intent(task):
    task = task.lower()

    if "email" in task:
        return "email"
    elif "summary" in task or "summarize" in task:
        return "summary"
    elif "code" in task:
        return "code"
    else:
        return "general"


def ai_agent(task):
    intent = detect_intent(task)

    if intent == "email":
        return "📧 Email Generated: Hello, I hope you're doing well. Let’s connect soon."

    elif intent == "summary":
        return "📝 Summary: This content has been summarized using AI logic."

    elif intent == "code":
        return "💻 Code: def hello(): print('Hello World')"

    else:
        return "🤖 General Response: Task processed successfully."
```

```python
print("🤖 AI Agent Started")

user_input = input("Enter your task: ")
```

```python
result = ai_agent(user_input)
print("\nResult:\n")
print(result)
```

```python
tests = [
    "write an email to manager",
    "summarize this document",
    "give me python code",
    "hello AI"
]

for t in tests:
    print("\nInput:", t)
    print("Output:", ai_agent(t))
```
